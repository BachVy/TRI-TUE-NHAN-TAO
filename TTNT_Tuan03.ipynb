# A. THUẬT GIẢI MINIMAX ỨNG DỤNG VÀO BÀI TOÁN TICTACTOE

```
import copy
import math
import random
import numpy
```

Import các thư viện cần thiết

```
X = "X"
O = "O"
EMPTY = None
user = None
ai = None
```

Thiết lập các giá trị cơ bản dùng để xác định người chơi, máy và trạng thái ô trên bàn cờ trước khi trò chơi bắt đầu.

```
def initial_state():
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]
```

Khởi tạo bàn cờ rỗng kích thước 3x3.

```
def player(board):
    """
    Returns player who has the next turn on a board.
    """
    count = 0
    for i in board:
        for j in i:
            if j:
                count += 1
    if count % 2 != 0:
        return ai
    return user
```

**Đầu vào:** Hàm nhận vào một biến board là một mảng 2 chiều, đại diện cho lưới ô vuông của bàn cờ.

**1. Đếm ô đã đánh:**
- Sử dụng hai vòng lặp for lồng nhau để duyệt qua tất cả các ô trên bàn cờ.
- Kiểm tra xem ô hiện tại (j) có chứa giá trị nào không.
- Biến count được tăng lên mỗi khi tìm thấy một ô đã được đánh.

**2. Xác định lượt đi:**
- Người chơi bắt đầu là người đánh lượt lẻ và người chơi thứ hai là người đánh lượt chẵn.
- Nếu count là số lẻ thì trả về người chơi là ai.
- Nếu count là số chẵn thì trả về người chơi là user.

```
def actions(board):
    """
    Returns set of all possible actions (i, j) available on the board.
    """
    res = set()
    board_len = len(board)
    for i in range(board_len):
        for j in range(board_len):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res
```

**Mục đích:** Hàm actions(board) dùng để trả về một tập hợp (set) chứa tọa độ (i, j) của tất cả các ô mà người chơi tiếp theo có thể đặt quân cờ của mình vào.

**1. Duyệt bàn cờ:**
- Lấy kích thước của bàn cờ.
- Duyệt qua tất cả các ô trên bàn cờ, i đại diện cho chỉ số hàng (row) và j đại diện cho chỉ số cột.

**2. Kiểm tra ô trống:**
- Kiểm tra xem ô hiện tại có rỗng không.
- Nếu ô tại tọa độ (i, j) trống thì tọa độ đó được thêm vào res.

**Kết quả:** Hàm trả về tập hợp res chứa tất cả các tọa độ của các ô trống.

```
def result(board, action):
    curr_player = player(board)
    result_board = copy.deepcopy(board)
    (i, j) = action
    result_board[i][j] = curr_player
    return result_board
```

**Mục đích:** Hàm result(board, action) mô phỏng một nước đi và trả về bàn cờ mới mà không làm thay đổi bàn cờ gốc.

**1. Xác định người chơi hiện tại:** Gọi hàm player(board) (mà chúng ta đã phân tích trước đó) để xác định người chơi nào đang đến lượt đi.

**2. Tạo bản sao**

**3. Thực hiện nước đi:**
- Lấy tọa độ của nước đi.
- Tại tọa độ (i, j) trên bàn cờ sao chép, ô trống được đánh dấu bằng ký hiệu của người chơi hiện tại.

**Kết quả:** Hàm trả về result_board, tức là trạng thái bàn cờ mới sau khi người chơi đã thực hiện nước đi.

```
def get_horizontal_winner(board):
    # check horizontally
    winner_val = None
    board_len = len(board)
    for i in range(board_len):
        winner_val = board[i][0]
        for j in range(board_len):
            if board[i][j] != winner_val:
                winner_val = None
        if winner_val:
            return winner_val
    return winner_val
```

**Mục đích:** Kiểm tra hàng ngang tìm người thắng.

**1. Duyệt từng hàng:**
- Duyệt qua từng hàng i.
- Gán người chơi ở ô đầu tiên của hàng hiện tại là người chiến thắng của hàng đó.

**2. Duyệt qua từng ô trong hàng hiện tại:**
- Nếu có bất kỳ ô nào trong hàng không khớp với giá trị winner_val ban đầu hoặc nếu winner_val là ô trống thì hàng đó không thể tạo thành một đường thắng.
- Loại bỏ người đó.

**3. Xác nhận:**
- Sau khi kiểm tra xong một hàng, nếu winner_val vẫn giữ được giá trị ban đầu thì giá trị của người chơi thắng được trả về ngay lập tức.
- Nếu không tìm thấy hàng chiến thắng thì hàm trả về None.

```
def get_vertical_winner(board):
    # check vertically
    winner_val = None
    board_len = len(board)
    for i in range(board_len):
        winner_val = board[0][i]
        for j in range(board_len):
            if board[j][i] != winner_val:
                winner_val = None
        if winner_val:
            return winner_val
    return winner_val
```

**Mục đích:** Kiểm tra hàng dọc tìm người thắng.

**1. Duyệt từng cột:**
- Duyệt qua từng cột i.
- Gán người chơi ở ô đầu tiên của cột hiện tại là người chiến thắng.

**2. Duyệt qua từng ô trong cột hiện tại:**
- Nếu có bất kỳ ô nào trong cột không khớp với giá trị winner_val ban đầu hoặc nếu winner_val là ô trống thì cột đó không thể tạo thành một đường thắng.
- Loại bỏ người đó.

**3. Xác nhận:**
- Sau khi kiểm tra xong một cột, nếu winner_val vẫn giữ được giá trị ban đầu thì giá trị của người chơi thắng được trả về ngay lập tức.
- Nếu không tìm thấy cột chiến thắng thì hàm trả về None.

```
def get_diagonal_winner(board):
    # check diagonally
    winner_val = None
    board_len = len(board)
    winner_val = board[0][0]
    for i in range(board_len):
        if board[i][i] != winner_val:
            winner_val = None
    if winner_val:
        return winner_val
    winner_val = board[0][board_len - 1]
    for i in range(board_len):
        j = board_len - 1 - i
        if board[i][j] != winner_val:
            winner_val = None
    return winner_val
```

**Mục đích:** Kiểm tra hàng chéo tìm người thắng.

**1. Duyệt từng hàng chéo:**
- Duyệt qua từng hàng chéo i.
- Gán người chơi ở ô đầu tiên của hàng chéo hiện tại là người chiến thắng.

**2. Duyệt qua từng ô trong hàng chéo:**
- Nếu có bất kỳ ô nào trong hàng chéo không khớp với giá trị winner_val ban đầu hoặc nếu winner_val là ô trống thì hàng chéo đó không thể tạo thành một đường thắng.
- Loại bỏ người đó.

**3. Xác nhận:**
- Sau khi kiểm tra xong một hàng chéo, nếu winner_val vẫn giữ được giá trị ban đầu thì giá trị của người chơi thắng được trả về ngay lập tức.
- Nếu không tìm thấy hàng chéo chiến thắng thì hàm trả về None.

```
def winner(board):
    winner_val = get_horizontal_winner(board) or get_vertical_winner(board) or get_diagonal_winner(board) or None
    return winner_val
```

**Mục đích:** Kiểm tra và trả về người chơi thắng

- Kiểm tra hàng ngang. Nếu tìm thấy người thắng, giá trị đó được gán cho winner_val và kết thúc.
- Nếu không có người thắng hàng ngang, kiểm tra tiếp tục hàng dọc.
- Nếu không có người thắng hàng ngang hay cột dọc, tiếp tục kiểm tra đường chéo.
- Nếu cả ba hàm trên đều trả về None thì giá trị cuối cùng được gán cho winner_val là None.

```
def terminal(board):
    if winner(board) != None:
        return True
    for i in board:
        for j in i:
            if j == EMPTY:
                return False
    return True
```

**Mục đích:** Kiểm tra trạng thái, trả về True nếu trò chơi không thể tiếp tục và False nếu vẫn còn nước đi.

Kiểm tra điều kiện thắng:
- Gọi hàm winner(board) để kiểm tra có người chơi nào đã giành chiến thắng chưa. Nếu có người thắng thì hàm trả về True.
- Nếu không có người thắng, tiếp tục kiểm tra xem còn ô trống không. Nếu có thì hàm trả về False.
- Nếu hết vòng lặp mà không tìm thấy ô trống và không có ai thắng thì trả về True.


```
def utility(board):
    winner_val = winner(board)
    if winner_val == X:
        return 1
    elif winner_val == O:
        return -1
    return 0
```

**Mục đích:** Hàm đánh giá cho trạng thái bàn cờ khi trò chơi đã kết thúc.

**1. Xác định người chiến thắng của bàn cờ hiện tại**

**2. Gán Điểm số:**
- Nếu người chơi X thắng, hàm trả về 1.
- Nếu người chơi O thắng, hàm trả về -1.
- Nếu không có ai thắng, hàm trả về 0.

**Vai trò:** Cho phép thuật toán tính toán ngược từ các trạng thái kết thúc trở về trạng thái hiện tại để chọn nước đi tối ưu.

```
def maxValue(state):
    if terminal(state):
        return utility(state)
    v = -math.inf
    for action in actions(state):
        v = max(v, minValue(result(state, action)))
    return v
```

**Mục đích:** Tính toán giá trị tối đa mà người chơi có thể đạt được trong trò chơi.

- Xác định điều kiện dừng: Nếu trạng thái bàn cờ hiện tại là trạng thái kết thúc thì trả về utility của trạng thái đó.
- Lặp qua tất cả action từ trạng thái hiện tại.
- Tạo ra trạng thái bàn cờ mới sau khi thực hiện action.
- Gọi đệ quy hàm minValue để mô phỏng việc người chơi Max chọn một nước đi và sau đó Min sẽ chơi tối ưu trên trạng thái đó.
- Người chơi Max chọn giá trị lớn nhất trong số tất cả các giá trị mà đối thủ (Min) có thể trả lại.

**Kết quả:** trả về giá trị tiện ích tối đa mà người chơi Max có thể đạt được từ trạng thái hiện tại.

```
def minValue(state):
    if terminal(state):
        return utility(state)
    v = math.inf
    for action in actions(state):
        v = min(v, maxValue(result(state, action)))
    return v
```

**Mục đích:** Tính toán giá trị tối thiểu mà người chơi có thể đạt được trong trò chơi.

- Xác định điều kiện dừng: Nếu trạng thái bàn cờ hiện tại là trạng thái kết thúc thì trả về utility của trạng thái đó.
- Lặp qua tất cả action từ trạng thái hiện tại.
- Tạo ra trạng thái bàn cờ mới sau khi thực hiện action.
- Gọi đệ quy hàm maxValue để mô phỏng việc người chơi Min chọn một nước đi và sau đó đối thủ (Max) sẽ chơi tối ưu trên trạng thái đó.
- Người chơi Min chọn giá trị nhỏ nhất trong số tất cả các giá trị mà đối thủ (Max) có thể trả lại.

**Kết quả:** trả về giá trị tiện ích tối đa mà người chơi Min có thể đạt được từ trạng thái hiện tại.

```
def minimax(board):
    current_player = player(board)
    if current_player == X:
        max = -math.inf
        for action in actions(board):
            check = minValue(result(board, action))  # FIXED
            if check > max:
                max = check
                move = action
    else:
        min = math.inf
        for action in actions(board):
            check = maxValue(result(board, action))  # FIXED
            if check < min:
                min = check
                move = action
    return move
```

**Mục đích:** Xác định nước đi tốt nhất mà người chơi nên thực hiện tại trạng thái bàn cờ hiện tại.

**Xác định người chơi hiện tại:**

**1. Nếu là Max:**
- Khởi tạo giá trị tiện ích.
- Duyệt qua tất cả các nước đi hợp lệ.
- Đối với mỗi nước đi, hàm tính toán giá trị Minimax bằng cách gọi minValue trên trạng thái bàn cờ kết quả sau nước đi đó.
- Nếu giá trị check mà nước đi này mang lại lớn hơn giá trị tối đa đã tìm thấy (max) thì nước đi này là tốt hơn.
- Cập nhật giá trị tối đa mới.
- Lưu lại nước đi tối ưu tương ứng.

**2. Nếu là Min:**
- Khởi tạo giá trị tiện ích.
- Duyệt qua tất cả các nước đi hợp lệ.
- Đối với mỗi nước đi, hàm tính toán giá trị Minimax bằng cách gọi maxValue trên trạng thái bàn cờ kết quả sau nước đi đó.
- Nếu giá trị check mà nước đi này mang lại nhỏ hơn giá trị tối thiểu đã tìm thấy (min) thì nước đi này là tốt hơn.
- Cập nhật giá trị tối thiểu mới.
- Lưu lại nước đi tối ưu tương ứng.

**Kết quả:** Trả về move là tọa độ (i, j) của nước đi tốt nhất mà người chơi hiện tại nên thực hiện.

```
if __name__ == "__main__":
    board = initial_state()
    ai_turn = False
    print("Choose a player")
    user = input()
    if user == "X":
        ai = "O"
    else:
        ai = "X"
    while True:
        game_over = terminal(board)
        playr = player(board)
        if game_over:
            winner = winner(board)
            if winner is None:
                print("Game Over: Tie.")
            else:
                print(f"Game Over: {winner} wins.")
            break;
        else:
            if user != playr and not game_over:
                if ai_turn:
                    move = minimax(board)
                    board = result(board, move)
                    ai_turn = False
                    print(numpy.array(board))
            elif user == playr and not game_over:
                ai_turn = True
                print("Enter the position to move (row,col)")
                i = int(input("Row:"))
                j = int(input("Col:"))
                if board[i][j] == EMPTY:
                    board = result(board, (i, j))
                    print(numpy.array(board))
```

# B. THUẬT TOÁN CẮT TỈA ALPHA-BETA VỚI ỨNG DỤNG VÀO BÀI TOÁN TICTACTOE

```
import math
import os
```

Import thư viện

```
X = "X"
O = "O"
EMPTY_CELL_TOKEN = " " # Sử dụng khoảng trắng thay vì số để dễ hiển thị hơn
BOARD_SIZE = 9
```

Khởi tạo hằng số trò chơi

```
def GetWinner(board):
    """
    Returns the winner (X hoặc O) in the current board if there is one,
    otherwise returns None.
    Board: danh sách 9 phần tử (index 0 đến 8).
    """
    # Các tổ hợp chiến thắng (chỉ số của 3 ô liên tiếp)
    WIN_LINES = [
        # Ngang
        (0, 1, 2), (3, 4, 5), (6, 7, 8),
        # Dọc
        (0, 3, 6), (1, 4, 7), (2, 5, 8),
        # Chéo
        (0, 4, 8), (2, 4, 6)
    ]

    for a, b, c in WIN_LINES:
        if board[a] == board[b] == board[c] and board[a] != EMPTY_CELL_TOKEN:
            return board[a]

    return None
```

**Mục đích:** xác định trạng thái chiến thắng
1. **Khởi tạo các tổ hợp chiến thắng:** Mảng hằng số WIN_LINES là một danh sách chứa các bộ ba chỉ số (dưới dạng tuple), đại diện cho toàn bộ 8 đường thắng hợp lệ trên lưới chơi 3x3 (được biểu diễn như một mảng phẳng gồm 9 ô, từ chỉ số 0 đến 8).
- Hàng ngang: (0, 1, 2), (3, 4, 5), (6, 7, 8)
- Cột dọc: (0, 3, 6), (1, 4, 7), (2, 5, 8)
- Đường chéo: (0, 4, 8), (2, 4, 6)
2. **Duyệt và kiểm tra từng đường thắng:** Vòng lặp for a, b, c in WIN_LINES: duyệt qua từng bộ ba chỉ số chiến thắng.
- Kiểm tra xem ba ô trên bàn cờ tại các chỉ số $a, b, c$ có cùng một ký hiệu hay không.
- Đảm bảo rằng ba ô đó không phải là ô trống. Điều kiện này là cần thiết vì nếu ba ô trống có cùng giá trị(EMPTY_CELL_TOKEN), chúng sẽ thỏa mãn Phần 1, nhưng điều đó không được tính là một chiến thắng.

**Kết quả:** Trả về ký hiệu của người chơi thắng cuộc (board[a]) và kết thúc. Nếu vòng lặp kết thúc mà không tìm thấy bất kỳ đường thắng nào, hàm sẽ trả về None.

- Giả định rằng board là một danh sách (list) phẳng gồm 9 phần tử (từ board[0] đến board[8]), đại diện cho 9 ô trên bàn cờ.
- Sắp xếp 9 phần tử này thành 3 hàng và 3 cột, sử dụng ký tự gạch đứng (|) để phân cách các cột và gạch ngang (-) để phân cách các hàng.

```
def GetAvailableCells(board):
    available = []
    for i in range(BOARD_SIZE):
        if board[i] == EMPTY_CELL_TOKEN:
            available.append(i + 1)
    return available
```

**Mục đích:** cung cấp danh sách các nước đi hợp lệ (các ô trống) cho thuật toán.
1. **Khởi tạo danh sách:** Khởi tạo một danh sách trống để lưu trữ số thứ tự của các ô hợp lệ (ô trống) được tìm thấy.
2. **Duyệt qua bàn cờ:** duyệt qua tất cả các chỉ số (index) của bàn cờ, từ 0 đến 8.
3. **Kiểm tra trạng thái ô**: Kiểm tra giá trị của ô tại chỉ số i. Hàm chỉ quan tâm các ô có giá trị bằng EMPTY_CELL_TOKEN.
4. **Thêm nước đi Hợp lệ:** Nếu ô đó là trống, hàm tính toán số thứ tự của ô bằng cách lấy chỉ số mảng và cộng thêm 1.

**Kết quả:** Trả về danh sách chứa số thứ tự của tất cả các ô hiện đang trống trên bàn cờ.

```
def minimax(position, depth, alpha, beta, isMaximizing):
    winner = GetWinner(position)

    # 1. Điều kiện Dừng (Terminal Check)
    if winner != None:
        # X thắng (Maximizer) -> Giá trị dương, ưu tiên thắng nhanh
        if winner == X:
            return 10 - depth
        # O thắng (Minimizer) -> Giá trị âm, ưu tiên thua chậm
        elif winner == O:
            return -10 + depth

    # Hòa
    if len(GetAvailableCells(position)) == 0:
        return 0

    # 2. Người chơi Max (X)
    if isMaximizing:
        maxEval = -math.inf

        # Lặp qua các nước đi khả thi
        for cell_num in GetAvailableCells(position):
            cell_index = cell_num - 1

            # Thực hiện nước đi và gọi đệ quy
            position[cell_index] = X
            Eval = minimax(position, depth + 1, alpha, beta, False)

            # Hoàn tác nước đi (Unmake move)
            position[cell_index] = EMPTY_CELL_TOKEN

            maxEval = max(maxEval, Eval)
            alpha = max(alpha, Eval)

            # Cắt tỉa (Pruning)
            if beta <= alpha:
                break
        return maxEval

    # 3. Người chơi Min (O)
    else:
        minEval = math.inf

        # Lặp qua các nước đi khả thi
        for cell_num in GetAvailableCells(position):
            cell_index = cell_num - 1

            # Thực hiện nước đi và gọi đệ quy
            position[cell_index] = O
            Eval = minimax(position, depth + 1, alpha, beta, True)

            # Hoàn tác nước đi (Unmake move)
            position[cell_index] = EMPTY_CELL_TOKEN

            minEval = min(minEval, Eval)
            beta = min(beta, Eval)

            # Cắt tỉa (Pruning)
            if beta <= alpha:
                break
        return minEval
```

**Mục đích:** Đánh giá tất cả các nước đi tương lai có thể xảy ra từ trạng thái hiện tại để xác định xem trạng thái này cuối cùng sẽ dẫn đến chiến thắng, thất bại hay hòa và gán một giá trị số cho kết quả đó.

**Tham số:**
- **position:** Trạng thái bàn cờ hiện tại.
- **depth:** Độ sâu hiện tại trong cây tìm kiếm (số nước đã đi từ trạng thái gốc).
- **alpha:**	Giá trị tốt nhất đã đảm bảo cho người chơi MAX (tối đa hóa).
- **beta:**	Giá trị tốt nhất đã đảm bảo cho người chơi MIN (tối thiểu hóa).
- **isMaximizingTrue:** Nếu là lượt của người chơi MAX (X), False nếu là lượt của người chơi MIN (O).

1. **Điều kiện dừng (Tính điểm tiện ích):**
- Nếu X thắng (Maximizer): Trả về 10 - depth. (Giá trị dương: tốt cho X. Trừ depth để ưu tiên thắng nhanh hơn).
- Nếu O thắng (Minimizer): Trả về -10 + depth. (Giá trị âm: tốt cho O. Cộng depth để ưu tiên thua chậm hơn).
- Nếu không còn ô trống trả về 0 (Hòa).
2. **Lượt Chơi Max:**
- Đặt giá trị tối đa tìm được ban đầu nhỏ.
- Lặp qua tất cả các nước đi hợp lệ.
- Gọi đệ quy cho lượt của người chơi MIN trên trạng thái bàn cờ mới.
- Chọn giá trị tối đa từ các nước đi có thể xảy ra.
- Cập nhật giá trị Alpha (giá trị tốt nhất mà MAX có thể đảm bảo).
- Nếu giá trị đảm bảo của MIN (beta) nhỏ hơn hoặc bằng giá trị đảm bảo của MAX (alpha), thì nhánh này không bao giờ được MIN chọn, nên ta cắt tỉa và dừng tìm kiếm nhánh này.
- Khôi phục trạng thái bàn cờ sau khi đánh giá xong một nhánh để không ảnh hưởng đến các lần lặp tiếp theo.
3. **Lượt Chơi Min:**
- Đặt giá trị tối thiểu tìm được ban đầu lớn.
- Lặp qua tất cả các nước đi hợp lệ.
- Gọi đệ quy cho lượt của người chơi MAX trên trạng thái bàn cờ mới.
- Chọn giá trị tối thiểu từ các nước đi có thể xảy ra.
- Cập nhật giá trị beta (giá trị tốt nhất mà MIN có thể đảm bảo).
- Nếu giá trị đảm bảo của MIN (beta) nhỏ hơn hoặc bằng giá trị đảm bảo của MAX (alpha), thì cắt tỉa.
- Khôi phục trạng thái bàn cờ.

```
def FindBestMove(currentPosition, AI):
    """
    Sử dụng minimax để tìm nước đi (chỉ số ô 1-9) tốt nhất cho AI.
    """
    bestMove = -1

    # AI là X (Maximizer)
    if AI == X:
        bestVal = -math.inf
        isMaximizing = False # Người chơi tiếp theo là MIN

        for cell_num in GetAvailableCells(currentPosition):
            cell_index = cell_num - 1

            # Đánh giá nước đi
            currentPosition[cell_index] = AI
            moveVal = minimax(currentPosition, 0, -math.inf, math.inf, isMaximizing)
            currentPosition[cell_index] = EMPTY_CELL_TOKEN # Hoàn tác

            if moveVal > bestVal:
                bestVal = moveVal
                bestMove = cell_num

    # AI là O (Minimizer)
    else:
        bestVal = math.inf
        isMaximizing = True # Người chơi tiếp theo là MAX

        for cell_num in GetAvailableCells(currentPosition):
            cell_index = cell_num - 1

            # Đánh giá nước đi
            currentPosition[cell_index] = AI
            moveVal = minimax(currentPosition, 0, -math.inf, math.inf, isMaximizing)
            currentPosition[cell_index] = EMPTY_CELL_TOKEN # Hoàn tác

            if moveVal < bestVal:
                bestVal = moveVal
                bestMove = cell_num

    return bestMove
```

**Mục đích:** Lựa chọn nước đi thực tế tối ưu nhất cho người chơi AI tại trạng thái bàn cờ hiện tại, dựa trên kết quả đánh giá điểm tiện ích (utility score) được tính toán bởi thuật toán Minimax.
1. **Áp dụng cho cả hai trường hợp**
- Duyệt qua từng ô trống (được trả về dưới dạng số ô 1-9) trên bàn cờ.
- Giả định AI thực hiện nước đi vào ô trống hiện tại.
- Gọi hàm minimax để tính toán giá trị tiện ích Minimax của trạng thái bàn cờ mới này. Độ sâu khởi đầu là 0.
- Khôi phục bàn cờ về trạng thái trước khi đánh giá để tiếp tục đánh giá các nước đi khác.
- Lưu lại nước đi cho ra giá trị tốt nhất tìm được cho đến thời điểm hiện tại.
2. **Trường hợp AI là X (Maximizer)**
- Khởi tạo: bestVal được đặt rất nhỏ vì AI (X) muốn tối đa hóa giá trị tiện ích (tìm kiếm giá trị +10).
- Tham số đệ quy: isMaximizing = False được truyền vào minimax. Điều này là chính xác vì sau khi AI (X) đi, người chơi tiếp theo là đối thủ (O), tức là người chơi Minimizing.
- AI chọn nước đi có giá trị moveVal lớn hơn giá trị tốt nhất hiện tại.
3. **Trường hợp AI là O (Minimizer)**
- Khởi tạo: bestVal được đặt rất lớn vì AI (O) muốn tối đa hóa giá trị tiện ích (tìm kiếm giá trị -10).
- Tham số đệ quy: isMaximizing = True được truyền vào minimax. Điều này là chính xác vì sau khi AI (O) đi, người chơi tiếp theo là đối thủ (O), tức là người chơi Maximizing.
- AI chọn nước đi có giá trị moveVal nhỏ hơn giá trị tốt nhất hiện tại.

```
def main():
    # 1. Thiết lập trò chơi
    while True:
        player_char = input("Play as X or O? ").strip().upper()
        if player_char in [X, O]:
            break
        print("Invalid choice. Please enter 'X' or 'O'.")

    AI = O if player_char == X else X

    currentGame = [EMPTY_CELL_TOKEN] * BOARD_SIZE

    # X luôn bắt đầu trước
    currentTurn = X

    print(f"\nYou are {player_char}, AI is {AI}. {X} starts first.")

    # 2. Vòng lặp trò chơi
    while True:
        # 2a. Xử lý lượt chơi
        if currentTurn == AI:
            print("\nAI is making a move...")
            # Sử dụng FindBestMove để tìm nước đi
            cell_num = FindBestMove(currentGame, AI)
            cell_index = cell_num - 1
            currentGame[cell_index] = AI
            currentTurn = player_char

        elif currentTurn == player_char:
            PrintBoard(currentGame)
            while True:
                try:
                    humanInput = int(input("Enter cell number (1-9): ").strip())
                    cell_index = humanInput - 1

                    if 1 <= humanInput <= BOARD_SIZE and currentGame[cell_index] == EMPTY_CELL_TOKEN:
                        currentGame[cell_index] = player_char
                        currentTurn = AI
                        break
                    else:
                        print("Invalid move. Cell not available or input out of range (1-9).")
                except ValueError:
                    print("Invalid input. Please enter a number.")

        # 2b. Kiểm tra kết thúc trò chơi
        winner = GetWinner(currentGame)
        if winner != None:
            PrintBoard(currentGame)
            print(f"\n--- GAME OVER: {winner} WON!!! ---")
            break

        if len(GetAvailableCells(currentGame)) == 0:
            PrintBoard(currentGame)
            print("\n--- GAME OVER: Tie. ---")
            break
```

**Mục đích:** Khởi tạo và chạy vòng lặp trò chơi cho đến khi đạt được trạng thái kết thúc (thắng, thua, hoặc hòa).

### Thiết lập trò chơi
1. Chọn vai trò (X hoặc O) và gán vai trò còn lại cho AI.
2. Tạo bàn cờ trống ban đầu (danh sách 9 phần tử).
3. Đặt người chơi X luôn là người đi trước.
4. Thông báo cho người dùng biết vai trò của họ và ai bắt đầu.
### Vòng lặp chính
1. Gọi hàm FindBestMove(currentGame, AI) để nhận nước đi tối ưu (sử dụng Minimax), cập nhật bàn cờ và chuyển lượt cho người chơi.
2. Người dùng gọi PrintBoard(currentGame) để hiển thị trạng thái hiện tại.
3. Vòng lặp while True: bên trong yêu cầu người dùng nhập số ô (1-9).
4. Kiểm tra xem ô đó có hợp lệ không (trong phạm vi 1-9 và còn trống).
5. Nếu hợp lệ, cập nhật bàn cờ và chuyển lượt cho AI; nếu không, thông báo lỗi và yêu cầu nhập lại.
6. Sau mỗi nước đi (của AI hoặc người dùng), chương trình kiểm tra hai điều kiện dừng:
- **Kiểm tra thắng thua:** Nếu winner != None, chương trình in bàn cờ, thông báo người chiến thắng và thoát vòng lặp (break).
- **Kiểm tra hòa:** Nếu không còn ô trống nào (và chưa có người thắng), chương trình in bàn cờ, thông báo Hòa (Tie.) và thoát vòng lặp (break).
